**Adding And Making Layers of our ANN**

```
# Now importing the models for ANN
import keras
# For make ANN step by  sequential is used for that model which we use for classifying
from keras import Sequential
# Adding layers through Dense
from keras.layers import Dense
```

```
# Adding the Input and First Hidden Layers means neurons layer

classifier = Sequential()

# Now adding layers one by one

value = int(len(df.iloc[:, 2:].columns)) # no of input variables

classifier.add(
    # First of all specify number of independent variables (input_dim)
    # Now second parameter we add no of hidden layers by using this technique round(number of columns / 2)
    # Third parameter is show what is the value of weight on layer i use uniform distribution which decides values nears to zero
    # Fouth parameter is a activation function as we now in hidden layer we use relu function so i use relu function
    Dense( 
          round(int(len(df.columns)) / 2),
          kernel_initializer='random_uniform',
          activation = 'relu',
          input_dim = 30
        )
)
# Finally i add input layer with hidden layer having 16 hidden layers



# Adding second hidden layer

classifier.add(
    # Now second parameter we add no of hidden layers by using this technique round(number of columns / 2)
    # Third parameter is show what is the value of weight on layer i use uniform distribution which decides values nears to zero
    # Fouth parameter is a activation function as we now in hidden layer we use relu function so i use relu function
    Dense( 
          round(int(len(df.columns)) / 2),
          kernel_initializer='random_uniform',
          activation = 'relu'
        )
)


# Adding Output Layer

classifier.add(
    # I use 1 because at output layer there is only one output
    # Third parameter is show what is the value of weight on layer i use uniform distribution which decides values nears to zero
    # Fouth parameter is a activation function as we now in output layer we use sigmoid function so i use relu function
    Dense( 
          1,
          kernel_initializer='random_uniform',
          activation = 'sigmoid'
        )
)
```

**After Making Layers of ANN**

```
classifier.compile(optimizer="Adam", loss='binary_crossentropy', metrics=['accuracy'])
# Loss function help in backpropogration in ANN
# if your output in binary then we use a loss function called binary_crossentropy
# if we have multiple values then we use categorical_crossentropy
# Adam is a type of Stocastic Gradient Descent
# metrices in used to show accuracy during every epoch
# epoch is nothing but a whole training on dataset in one time
```

**Training Start after designing neural network**

```
# Now train neural network
# First add variables on which we want to train our ANN
# batch_size is a paramter which determines how much a ANN select data at one time
# Epoch paramter means number of times ANN run whole training
# Batch size should be less i think after experiment

# Mini Batch Gradient Descent
classifier.fit(x_train, y_train, batch_size = 100, epochs=150)
```

**K-Fold Cross Validation(A Step To Increase Accuracy)**

```
# Import libraries
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score,cross_val_predict
from keras.models import Sequential 
from keras.layers import Dense
```

**Make Classifier**

```
def built_classifier():
  classifier = Sequential()
  value = int(len(df.iloc[:, 2:].columns)) # no of input variables
  classifier.add(
      Dense( 
            round(int(len(df.columns)) / 2),
            kernel_initializer='random_uniform',
            activation = 'relu',
            input_dim = 30
          )
  )
  classifier.add(
      Dense( 
            round(int(len(df.columns)) / 2),
            kernel_initializer='random_uniform',
            activation = 'relu'
          )
  )

  classifier.add(
      Dense( 
            1,
            kernel_initializer='random_uniform',
            activation = 'sigmoid'
          )
  )
  classifier.compile(optimizer="Adam", loss='binary_crossentropy', metrics=['accuracy'])
  return classifier
```

**Now Main Part Starts here**

```
# First parameter is function name that contains classifier
classifier = KerasClassifier(build_fn = built_classifier, batch_size = 100, epochs = 150)

# Getting Acuracies
# cv means k-folds size
# n_jobs is for large datasets
# X means independent variables
# y means dependent variables
accuracies = cross_val_score(estimator=classifier, X = x_train, y=y_train, cv = 10, n_jobs = -1)
y_pred = cross_val_predict(estimator=classifier, X = x_train, y=y_train, cv = 10, n_jobs = -1)
```

```
# getting mean of accuracies

mean = accuracies.mean()
```

```
print("The mean of our model is",mean)
```

```
print("The variance of our model is",accuracies.std())
```

```
print("The Accuracy of our model is {}% accurate.".format(round(mean+accuracies.std(),3)))
```
